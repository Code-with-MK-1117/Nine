# Data Visualization

```python
plt.figure(figsize=(10,5))    
 

sns.countplot(data["CATEGORY"])     
plt.xticks(rotation = 90)            
plt.yticks(rotation = 90)          

plt.show
```

```python
plt.figure(figsize=(20,5))


plt.subplot(1,2,2)

sns.countplot(data["PURPOSE"])
plt.title("Trip Purpose Distribution")
plt.show()
```

```python
fig, ax = plt.subplots() 

ax = sns.countplot(x=data["DAY-NIGHT"], ax=ax)

ax.bar_label(ax.containers[0], fmt='%d', fontsize=12, fontweight='bold')

plt.show()
```

```python
data["month"] = pd.DatetimeIndex(data["START_DATE"]).month

month_label = { 1.0 : "January", 2.0 : "February",3.0 : "March", 4.0 : "April", 5.0 :"May", 6.0 :"June",
               7.0 : "July", 8.0 : "August", 9.0 :"September", 10.0 :"October", 11.0 :"November", 12.0 :"December" }

data["month"] = data.month.map (month_label)

mon = data.month.value_counts(sort=False)
```

```python
data.head(1)
```

```python
df = pd.DataFrame({
    "months": mon.index,   
    "Value Count": data.groupby('month', sort=False)["MILES"].max().values  
})
 
df["months"] = pd.Categorical(df["months"], categories=mon.index, ordered=True)

plt.figure(figsize=(10,5))
p = sns.lineplot(data=df, x="months", y="Value Count", marker="o")
p.set(xlabel="Months", ylabel="Max Miles")
plt.xticks(rotation=90)
plt.show()
```

```python
df = pd.DataFrame({
    "months": mon.index,   
    "Trip Count": mon.values,   
    "Max Miles": data.groupby('month', sort=False)["MILES"].max().values   
})

df["months"] = pd.Categorical(df["months"], categories=mon.index, ordered=True)

plt.figure(figsize=(12,6))
sns.lineplot(data=df, x="months", y="Trip Count", marker="o", label="Trip Count", color="b")
sns.lineplot(data=df, x="months", y="Max Miles", marker="o", label="Max Miles", color="r")

plt.xlabel("Months")
plt.ylabel("Count / Max Miles")
plt.title("Monthly Trip Count vs. Max Miles")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
data['DAY'] = data.START_DATE.dt.weekday

day_label = {
    0: 'mon', 1: ' Tues' , 3: 'Wed' , 4: 'Thur' , 5: 'Sat' , 6:'Sun'}
data['DAY'] = data['DAY'].map(day_label)
```
